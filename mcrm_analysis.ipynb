```python
import sympy as sp
```

$$\frac{dY}{dt} = Y \bigg[w_O c_O O + [w_S^{ana}​+(w_S^{aer}​−w_S^{ana}​)f(O)]u_S(S) - m\bigg]$$

$$\frac{dO}{dt} = -c_O O Y$$

$$\frac{dS}{dt} = r_S S (1 - \frac{S}{P_S}) - c_S Y S$$

$$\frac{dC}{dt} = (1 - f(O))u_S(S)Y * 0.0111$$

```python
Y, O, S, C = sp.symbols('Y O S C')

wo, waer, wana = sp.symbols('w_O w_S^{aer} w_S^{ana}')

wo = sp.Symbol('wo')
waer = sp.Symbol('waer')
wana = sp.Symbol('wana')

ko, ks = sp.symbols('k_O k_S')

ps, rs = sp.symbols('P_S r_S')

co, cs, m = sp.symbols('c_O c_S m')
```

```python
dxdt = sp.Matrix([[Y * (wo * co * O + (wana + (waer - wana) * (O / (ko + O))) * (cs * S / (ks + S))) - m],
                  [-co * O * Y],
                  [rs * S * (1 - (S / ps)) - cs * Y * S]])

J = dxdt.jacobian([Y, O, S])
J
```

```python
subs_dict = {wo: 0.0156,
             waer: .45,
             wana: .1,
             ko: .015,
             ks: .3,
             ps: 15,
             rs: .4,
             co: .5,
             cs: .15,
             m: 0}

J_with_vals = J.subs(subs_dict)

J_with_vals

# eigs = J_with_vals.eigenvals()
# eigs
```
